```python
import cv2
from matplotlib import pyplot as plt
```

```python
# create figure
fig = plt.figure(figsize=(30, 30))

# setting values to rows and column variables
rows = 3
columns = 3

# reading images
Image1 = cv2.cvtColor(cv2.imread('image/CARS/car.4.jpg'), cv2.COLOR_BGR2RGB)
Image2 = cv2.cvtColor(cv2.imread('image/CATS/cat_3_30.png'), cv2.COLOR_BGR2RGB)
Image3 = cv2.cvtColor(cv2.imread('image/daisy/daisy_1.jpg'), cv2.COLOR_BGR2RGB)
Image4 = cv2.cvtColor(cv2.imread('image/dandelion/dandelion_2.jpg'), cv2.COLOR_BGR2RGB)
Image5 = cv2.cvtColor(cv2.imread('image/DOGS/dog_9_19.png'), cv2.COLOR_BGR2RGB)
Image6 = cv2.cvtColor(cv2.imread('image/FACES/00004.jpg'), cv2.COLOR_BGR2RGB)
Image7 = cv2.cvtColor(cv2.imread('image/roses/rose_1.jpg'), cv2.COLOR_BGR2RGB)
Image8 = cv2.cvtColor(cv2.imread('image/sunflowers/sunflower_1.jpg'), cv2.COLOR_BGR2RGB)
Image9 = cv2.cvtColor(cv2.imread('image/tulips/tulips_1.jpg'), cv2.COLOR_BGR2RGB)

# Adds a subplot at the 1st position
fig.add_subplot(rows, columns, 1)

# showing image
plt.imshow(Image1)
plt.axis('off')
plt.title("Car",fontsize = 50)

# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(Image2)
plt.axis('off')
plt.title("Cat",fontsize = 50)
  
# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(Image3)
plt.axis('off')
plt.title("Daisy",fontsize = 50)
  
# Adds a subplot at the 4th position
fig.add_subplot(rows, columns, 4)
  
# showing image
plt.imshow(Image4)
plt.axis('off')
plt.title("Dandelion",fontsize = 50)

# Adds a subplot at the 5th position
fig.add_subplot(rows, columns, 5)
  
# showing image
plt.imshow(Image5)
plt.axis('off')
plt.title("Dog",fontsize = 50)

# Adds a subplot at the 6th position
fig.add_subplot(rows, columns, 6)
  
# showing image
plt.imshow(Image6)
plt.axis('off')
plt.title("Face",fontsize = 50)

# Adds a subplot at the 7th position
fig.add_subplot(rows, columns, 7)
  
# showing image
plt.imshow(Image7)
plt.axis('off')
plt.title("Rose",fontsize = 50)

# Adds a subplot at the 8th position
fig.add_subplot(rows, columns, 8)
  
# showing image
plt.imshow(Image8)
plt.axis('off')
plt.title("Sunflower",fontsize = 50)

# Adds a subplot at the 9th position
fig.add_subplot(rows, columns, 9)
  
# showing image
plt.imshow(Image9)
plt.axis('off')
plt.title("Tulips",fontsize = 50)

plt.show()
```
