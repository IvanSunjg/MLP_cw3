```python
with open("original_alexnet.txt","r",encoding="utf-8") as f:
    alexnet = f.read()

with open("pretrained_alexnet.txt","r",encoding="utf-8") as f:
    pre_alexnet = f.read()

with open("sub_alexnet.txt","r",encoding="utf-8") as f:
    sub_alexnet = f.read()

with open("pretrained_sub_alexnet.txt","r",encoding="utf-8") as f:
    pre_sub_alexnet = f.read()
```

```python
import re
plot_data = {"TL":[],"VL":[],"TA":[],"VA":[]}
for data in [alexnet,pre_alexnet,sub_alexnet,pre_sub_alexnet]:
    tl = re.findall(r"Train_Loss: ([0-9]\.[0-9]*) ",data)
    tl = [float(x) for x in tl]
    plot_data["TL"].append(tl)
    ta = re.findall(r"Train_Acc: ([0-9]\.[0-9]*)",data)
    ta = [float(x) for x in ta]
    plot_data["TA"].append(ta)
    vl = re.findall(r"Test_Loss: ([0-9]\.[0-9]*) ",data)
    vl = [float(x) for x in vl]
    plot_data["VL"].append(vl)
    va = re.findall(r"Test_Acc: ([0-9]\.[0-9]*)",data)
    va = [float(x) for x in va]
    plot_data["VA"].append(va)
```

```python
import matplotlib.pyplot as plt
plt.plot(plot_data["VL"][0],"r",label="OriginalAlexnet")
plt.plot(plot_data["VL"][1],"m",label="PretrainedAlexnet")
plt.plot(plot_data["VL"][2],"b",label="SubAlexNet")
plt.plot(plot_data["VL"][3],"c",label="PreSubAlexNet")
plt.xlabel("epoch")
plt.ylabel("loss")
plt.legend()
plt.show()
```

```python
plt.plot(plot_data["VA"][0],"r",label="OriginalAlexnet")
plt.plot(plot_data["VA"][1],"m",label="PretrainedAlexnet")
plt.plot(plot_data["VA"][2],"b",label="SubAlexNet")
plt.plot(plot_data["VA"][3],"c",label="PreSubAlexNet")
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```
